```python
%matplotlib inline
import sys
from types import ModuleType
from importlib import reload
```

```python
SRC_PATH = "./src"
sys.path.append(SRC_PATH)
import main
```

```python
def rreload(module):
    """Recursively reload modules."""
    reload(module)
    for attribute_name in dir(module):
        attribute = getattr(module, attribute_name)
        if type(attribute) is ModuleType and attribute.__file__.startswith(SRC_PATH):
            rreload(attribute)
```

```python
rreload(main)
main.main()
```
